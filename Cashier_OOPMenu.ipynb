```python
class Transaction:

  LOWER_AMOUNT = 200_000
  LOWER_DISCOUNT = 0.05
  MID_AMOUNT = 300_000
  MID_DISCOUNT = 0.08
  HIGH_AMOUNT = 500_000
  HIGH_DISCOUNT = 0.1
  TABLE_HEADERS = [
    "Item Name", "Item Quantity", "Item Unit", "Price/Item", "Total"
  ]
    
  def __init__(self):
    self.basket = []
    self.basket_item = ()

  def getPricelist(self):
    pricelist={'mangga_butir':('mangga_butir',10_000,'butir')
               ,'mangga_kilo':('mangga_kilo',25_000,'kg')
               ,'jambu_butir':('jambu_butir',2_000,'butir')
               ,'jambu_kilo':('jambu_kilo',25_000,'kg')
               ,'rambutan_butir':('rambutan_butir',1_500,'butir')
               ,'rambutan_kilo':('rambutan_kilo',27_500,'kg')}
    
    '''
               ,'duku_butir':('mobil',1_000,'butir')
               ,'duku_kilo':('duku_kilo',25_000,'kg')
               ,'chery_butir':('chery_butir',2_000,'butir')
               ,'chery_pak':('chery_pak',40_000,'pak')
               ,'alpukat_butir':('alpukat_butir',12_500,'butir')
               ,'alpukat_kilo':('alpukat_kilo',12_500,'kg')
               ,'semangka_butir':('semangka_butir',25_000,'butir')}
    '''
    return pricelist
    
  def new_basket_record(self,item_name, param_qty=0,param_paid=0):

    item_name=item_name.strip()
    item_qty=int(param_qty)
    item_paid=int(param_paid)

    pricelist=self.getPricelist()
    basket_record=list()

    #print('item_name :_',item_name, '\n'
    #      , 'param_qty :_',param_qty, '\n'
    #      , 'param_paid :_',param_paid, '\n') 
    
    item_price=0
    if (item_qty==0) : item_qty=1  
    item_price=pricelist[item_name][1]*item_qty
    
    if (item_paid>0) and (item_paid<item_price):    
      item_qty=int(item_paid/ pricelist[item_name][1])   
      item_price=pricelist[item_name][1]*item_qty

    basket_record=(pricelist[item_name][0]
                   , item_qty
                   , pricelist[item_name][2]
                   , pricelist[item_name][1]
                   , item_price)

    return basket_record  
    
  def add_item(self, item_name, item_qty=1, item_price=0):

    new_item=item_name.strip()
    
    basket_item=self.new_basket_record(new_item, item_qty, item_price)
    self.basket.append(basket_item)

    msg='\n'+ 'One item added to Basket'
    return msg
    
    
  def update_item_name(self, item_name, new_item_name):

    old_item=item_name.strip()
    new_item=new_item_name.strip()
    
    iFound=0  
    for row in range(len(self.basket)):

      if self.basket[row][0]==old_item: 
        iFound+=1

        basket_out=self.basket.pop(row)
        old_qty=basket_out[1]

        basket_item=self.new_basket_record(new_item,old_qty)
            
        self.basket.append(basket_item)
        msg=': One item in Basket is updated'
        break
            
      else:        
        msg=': Item is not found in Basket.'

    msg='\n'+ 'Search failed' + msg if (not iFound) else 'Search success' + msg
    return msg
    
    
  def update_item_qty(self, item_name, new_item_qty):
    
    iFound=0    
    for row in range(len(self.basket)):
        
      if self.basket[row][0]==item_name.strip(): 
        iFound+=1
        # print (f'found at {row}: {self.basket[row][0]}')

        basket_out=self.basket.pop(row)
        new_qty=new_item_qty
        basket_item=self.new_basket_record(basket_out[0],new_qty)
            
        self.basket.append(basket_item)
        msg=': One item in Basket is updated'
        break
            
      else:        
        msg=': Item is not found in Basket.'

    msg='\n'+ 'Search failed' + msg if (not iFound) else 'Search success' + msg
    return msg
    
   
  def update_item_price(self, item_name, new_item_price):
    
    iFound=0    
    for row in range(len(self.basket)):
        
      if self.basket[row][0]==item_name.strip(): 
        iFound+=1
        #print (f'found at {row}: {self.basket[row][0]}')
                
        basket_out=self.basket.pop(row)
        new_paid=int(new_item_price)
            
        if new_paid<basket_out[2]:
          basket_item=self.new_basket_record(basket_out[0],basket_out[1],new_paid)
        else:
          basket_item=self.new_basket_record(basket_out[0],basket_out[1],basket_out[2])
            
        self.basket.append(basket_item)
        msg=': One item in Basket is updated'
        break
            
      else:        
        msg=': Item is not found in Basket.'

    msg='\n'+ 'Search failed' + msg if (not iFound) else 'Search success' + msg
    return msg
    
  def delete_item(self, item_name):
    
    iFound=0
    for row in range(len(self.basket)):    
        
      if self.basket[row][0]==item_name.strip(): 
        iFound+=1
        # print (f'found at {row}: {self.basket[row][0]}')
                
        basket_out=self.basket.pop(row)

        msg='One item in Basket is deleted'
        break
            
      else:        
        msg=': Item is not found in Basket.'

    msg='\n'+ 'Search failed' + msg if (not iFound) else 'Search success'
    return msg
    
  def reset_transaction(self):
    # empty the basket, end the loop. 
    self.basket.clear()
    msg='\n'+ 'All transaction is deleted. Basket is empty now.'
    return msg
    
  def check_order(self):
    
    iFound=0
    iTotal=0
    basketlist=list()
    
    # show the basket. 
    for row in range(len(self.basket)):                    
            
      iFound+=1
      basket_item = [self.basket[row][0]
                    ,self.basket[row][1]
                    ,self.basket[row][3]
                    ,self.basket[row][4]]
      basketlist.append(basket_item)
      iTotal= iTotal + int(self.basket[row][3])
    
    if iFound==0:
      msg='Check order failed: No item is found in Basket.' 

    else: 
      #print(basketlist,'\n')          
      headers = ["Item Name", "Item Quantity", "Item Price", "Total"]
      print(tabulate(basketlist, headers=headers, tablefmt='psql'),'\n')   
   
      msg='\n'+ 'Check order successful: ' + str(iFound) + ' item(s) found.'+ '\n'
    return msg
            
  def total_price(self):
    
    iFound=0
    iTotal=0
    basketlist = list()
    
    iFound=len(self.basket)
    
    if iFound==0:
      msg='Check order failed: No item is found in Basket.' 
    
    else: 
      iFound=0
      iDiscount=0
      # show the basket with total. 
      for row in range(len(self.basket)):                    
            
        iFound+=1
        basket_item = [self.basket[row][0]
                      ,self.basket[row][1]
                      ,self.basket[row][3]
                      ,self.basket[row][4]]
        basketlist.append(basket_item)
        iTotal= iTotal + int(self.basket[row][4])
        
      if (iTotal>=self.HIGH_AMOUNT):
        iDiscPct=self.HIGH_DISCOUNT  *100
        #print('iDiscPct : ', iDiscPct, 'self.HIGH_DISCOUNT : ',self.HIGH_DISCOUNT)
        iDiscount=iTotal*self.HIGH_DISCOUNT
        iTotal=iTotal-iDiscount
      elif (iTotal>=self.MID_AMOUNT):
        iDiscPct=self.MID_DISCOUNT  *100
        #print('iDiscPct : ', iDiscPct, 'self.MID_DISCOUNT : ',self.MID_AMOUNT)
        iDiscount=iTotal*self.MID_DISCOUNT
        iTotal=iTotal-iDiscount
      elif (iTotal>=self.LOWER_AMOUNT):
        iDiscPct=self.LOWER_DISCOUNT  *100
        #print('iDiscPct : ', iDiscPct, 'self.LOWER_DISCOUNT : ',self.LOWER_DISCOUNT)
        iDiscount=iTotal*self.LOWER_DISCOUNT
        iTotal=iTotal-iDiscount
      else:
        iDiscPct=0
 
      headers = ["Item Name", "Item Quantity", "Item Price", "Total"]
      print(tabulate(basketlist, headers=headers, tablefmt='psql'),'\n')
    
      if iDiscPct==0: 
        msg='\n'+ 'Check Out successful: \n' + \
        'total discount: 0 \n' + 'total price: ' + str(iTotal) + '\n'
      else:
        msg='\n'+ 'Check Out successful: \n' + \
        'transaction discount: ' + str(iDiscPct) + '% \n' + \
        'total discount: ' + str(iDiscount) + '\n' + \
        'total price: ' + str(iTotal) + '\n'
        
    return msg
    
if __name__ == "__main__":
    pass
```

```python
# Common Messages During User Interaction
MENU_NOT_AVAILABLE_MSG = "Menu is not Available, Please Select \
    the Available Menu"
SELECTION_NOT_INTEGER_MSG = "Please Enter Numbers between 1-7 only."
BACK_TO_MAIN_MENU_MSG = "\nBack to main Menu\n"


def add_new_item_menu(transaction_class, command):
    """
    Adding item to the transaction.
    The main menu will be cleared and the adding item menu is prompted.
    User will be asked to input the item name, item quantity and
    item price in order. When the input is finished
    it will call Transaction.add_item() method.
    
    Parameters
    ----------
    transaction_class : Class
       The Transaction Class object.
    command : str
        The command to clear the screen
    """

    # Clear the main menu
    os.system(command)

    PL = transaction_class.getPricelist()
    headers = ["Item Code", "Item Name", "Price/Item", "Item Unit"]
    #print(PL)
    print(tabulate([(k,) + v for k, v in PL.items()], headers=headers
                   , tablefmt='psql'),'\n')


    # Displaying the add item menu
    print("=================================================\n",
          "** Please add one item to the menu accordingly **\n",
          "=================================================\n")
    item_name = input("Enter the Item Name [String]: ")
    item_qty = input("Enter the Item Quantity [Integer]: ")
    item_price = input("Enter the Item Price [Integer]: ")

    # Calling the transaction_class.add_item method
    message = transaction_class.add_item(
                item_name=item_name,
                item_qty=item_qty,
                item_price=item_price)
    print(message,'\n')


def modify_item_menu(transaction_class, command):
    """
    Displaying transaction modification menu.
    The main menu will be cleared and then user will be prompted with menu
    to be selected with integer:
    - Menu 1 - Update Old Item Name with the new one
        User will enter the old item name to be replaced and the new item name
        then it will call Transaction.update_item_name() method and
        update the transaction.
    - Menu 2 - Update the current item quantity with the new quantity
        User will be asked to enter the item name of the quantity (qty)
        they want to update. Then it will call Transaction.update_item_qty()
        method and update the transaction.
    - Menu 3 - Update the current item price with the new item price
        User will be asked to enter the item name of the price they
        want to update along with the new price value. Then it will call
        Transaction.update_item_price() method and update the transaction
    - Menu 4 - Back to main menu
        User decide to not update any value in the transaction,
        and back to main menu.
    Parameters
    ----------
    transaction_class : Class
       The Transaction Class object.
    command : str
        The command to clear the
    Exceptions
    ----------
    ValueError:
        If the user enter string when they select the menu,
        the exception will be raised, then it will print warning message
        and reset is cancelled.
    """

    # Clear the main menu
    os.system(command)

    basketlist = transaction_class.basket
    headers = ["Item Name", "Item Quantity", "Item Unit", "Price/Item", "Total"] 
    #headers = ["Nama Barang", "Jumlah Unit", "Harga ", "Unit Satuan"]
    print(tabulate(basketlist, headers=headers, tablefmt='psql'),'\n')
    
    # Displaying the modify transaction menu
    print("================================\n",
          "** Modifying Transaction Menu **\n",
          "================================\n"
          "1. Update Item Name\n"
          "2. Update Item Quantity\n"
          "3. Update Item Price\n"
          "4. Back to Main Menu\n")

    try:
        choice = int(input("Select Menu [1-4]: "))
        if (choice == 1):
            # Prompt input the be submitted by user
            item_name = input("\nEnter The Item Name: ")
            new_item_name = input("Enter The New Item Name: ")

            # Call Transation.update_item_name() method
            message = transaction_class.update_item_name(
                item_name=item_name,
                new_item_name=new_item_name)
            print(message,'\n')
            
        elif (choice == 2):
            # Prompt input the be submitted by user
            qty_item_name = input("\nEnter The Item Name: ")
            new_item_qty = input("Enter The New Quantity of the Item: ")

            # Call Transation.update_item_qty() method
            message = transaction_class.update_item_qty(
                item_name=qty_item_name,
                new_item_qty=new_item_qty)
            print(message,'\n')
            
        elif (choice == 3):
            # Prompt input the be submitted by user
            price_item_name = input("\nEnter The Item Name: ")
            new_item_price = input("Enter The New Price of the Item: ")

            # Call Transation.update_item_price() method
            message = transaction_class.update_item_price(
                item_name=price_item_name,
                new_item_price=new_item_price)
            print(message,'\n')
            
        elif (choice == 4):
            print(BACK_TO_MAIN_MENU_MSG)
            
        else:
            print(MENU_NOT_AVAILABLE_MSG)
            
    except (ValueError):
        print(SELECTION_NOT_INTEGER_MSG)


def delete_item_menu(transaction_class, command):
    """Displaying the menu for deleting single item
    The function will ask user to enter the name of the item
    they want to delete. Then it will call Transaction.delete_item method
    to execute the deletion.
    Parameters
    ----------
    transaction : Class
       The Transaction Class object.
    command : str
        The command to be used to clear the CLI
    """

    # Clearing the main menu
    os.system(command)

    basketlist = transaction_class.basket
    headers = ["Item Name", "Item Quantity", "Item Unit", "Price/Item", "Total"] 
    print(tabulate(basketlist, headers=headers, tablefmt='psql'),'\n')

    # Displaying the delete menu
    print("================================\n"
          + "** Delete Single Item Menu    **\n"
          + "================================\n")
    item_name = input("Please Enter the Item Name to be Deleted: ")

    print(f"\nAre You Sure You want to delete '{item_name}' item?\n"
          + "1. Yes\n"
          + "2. No\n")

    try:
        choice = int(input("Select The Menu [1-2]: "))
        if (choice == 1):
            # Calling Transaction.delete_item() method
            message = transaction_class.delete_item(item_name)
            print(message)
            
        elif (choice == 2):
            print(BACK_TO_MAIN_MENU_MSG)
            
        else:
            print(MENU_NOT_AVAILABLE_MSG)
            
    except (ValueError):
        print(SELECTION_NOT_INTEGER_MSG)


def reset_menu(transaction_class):
    """Displaying the option whether user want to reset transaction or not.
    User will be asked to select whether to chose to delete 
    all item in transaction or not.
    Parameters
    ----------
    transaction_class : Class
       The Transaction Class object.
    Exceptions
    ----------
    ValueError:
        If the user enter string when they select the menu,
        the exception will be raised, then it will print warning message
        and reset is cancelled.
    """

    print("\nAre You Sure you Want to Delete all of the Transaction\n"
          "1. Yes\n"
          "2. No\n")

    try:
        choice = int(input("Select Menu [1-3]: "))
        if (choice == 1):
            message = transaction_class.reset_transaction()
            print(message)
            
        elif (choice == 2):
            print("\nDelete is Cancelled, Back to Main Menu")
            
        else:
            print(MENU_NOT_AVAILABLE_MSG)
    except (ValueError):
        print(SELECTION_NOT_INTEGER_MSG)


def main():
    """
    Displaying Application Main Menu.
    This function will instantiate Transaction class everytime the program run.
    This program will start by CLI-Based menu, where the user can select
    which action to run by typing the numbers listed in menu. 
    
    Exceptions
    ----------
    ValueError
        An exception will raise if a user enter other characters different 
        to the menu.
    """

    transact = Transaction()

    # Select Clear Command for the Command Line Interface based on the OS
    clear_command = "cls" if os.name == "nt" else "clear"

    # Main Menu For Our Program
    while True:
        print("=======================================\n"
              + "**   Cashier Application             **\n"
              + "=======================================\n"
              + "1. Add Item for Transaction\n"
              + "2. Modify Item in Transaction\n"
              + "3. Delete Item(s) in Transaction\n"
              + "4. Delete All of the Transactions\n"
              + "5. Check Transaction, Show list\n"
              + "6. Check Total Amount of Transaction\n"
              + "7. Exit From Program\n")

        try:
            choice = int(input("Select Menu [1-7]: "))
            if (choice == 1):
                add_new_item_menu(transact, clear_command)
            elif (choice == 2):
                modify_item_menu(transact, clear_command)
            elif (choice == 3):
                delete_item_menu(transact, clear_command)
            elif (choice == 4):
                reset_menu(transact)
            elif (choice == 5):
                message = transact.check_order()
                print(message,'\n')
            elif (choice == 6):
                message = transact.total_price()
                print(message,'\n')
            elif (choice == 7):
                exit()
                #quit()
            else:
                print(MENU_NOT_AVAILABLE_MSG)
                input("\nPress Enter to Continue\n")
                os.system(clear_command)
            
        except (ValueError):
            print(SELECTION_NOT_INTEGER_MSG)
            input("\nPress Enter to Continue\n")
            os.system(clear_command)

if __name__ == "__main__":
    main()
```
